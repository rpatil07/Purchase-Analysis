```python
import pandas as pd
import plotly.express as px 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

**Obective**:  Visual analysis of necessary purchases to luxury products based on income and age.

**Assumption**:
>necessary : meat,fruits and fish.

>luxury : wines,sweets and gold.

```python
df = pd.read_csv("https://raw.githubusercontent.com/nailson/ifood-data-business-analyst-test/master/ifood_df.csv")
```

**Purchases assumed as neccesary**

```python
plt.figure(figsize=(15,8))
ax = sns.scatterplot(data=df, x="Income", y="MntMeatProducts",hue="Age")
```

**Purchases assumed as luxury**

```python
plt.figure(figsize=(15,8))
ax = sns.scatterplot(data=df, x="Income", y="MntWines",hue="Age")
```

```python
plt.figure(figsize=(15,8))
ax = sns.scatterplot(data=df, x="Income", y="MntSweetProducts",hue="Age")
```

The purchase of sweets products are mostly from households with income 60000 onwards. There are however few cases where households with income 40000 also has purchased large amount of sweet products. The data shows that most of the products are purchased in higher quantity by households with more than 50000 income.

```python
plt.figure(figsize=(15,8))
ax = sns.scatterplot(data=df, x="Income", y="MntGoldProds",hue="Age")
```

Apart from some extreme cases, the purchase of gold products are mostly from households with income 40000 onwards.
