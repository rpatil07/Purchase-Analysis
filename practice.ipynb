```python
import pandas as pd
import plotly.express as px 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

**Obective**:  Visual analysis of necessary purchases to luxury products based on income and age.

**Assumption**:
>necessary : meat,fruits and fish.

>luxury : wines,sweets and gold.

```python
df = pd.read_csv("https://raw.githubusercontent.com/nailson/ifood-data-business-analyst-test/master/ifood_df.csv")
```

```python
sns.histplot(data=df, x="Income",bins = 10)
```

```python
counts, bins = np.histogram(df.Age, bins=range(20, 90, 5))
bins = 0.5 * (bins[:-1] + bins[1:])

fig = px.bar(x=bins, y=counts, labels={'x':'Age', 'y':'count'})
fig.show()
```
